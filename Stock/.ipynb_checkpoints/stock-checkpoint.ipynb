```python
import pandas as pd
```

```python
stock = pd.read_csv('assets/data/WMT.csv')
stock.head()
```

```python
Color=[]
Difference=[]
for i in range(len(stock)):
    Difference.append(stock["Close"][i]-stock["Open"][i])
    if Difference[i]>=0:
        Color.append("#8b0000")
    
    elif Difference[i]<0:
        Color.append("#006400")
stock["Difference"]=Difference
stock["Color"]=Color
stock.head()
```

```python
moving_avg= stock['Difference'].rolling(window=180).mean()
stock["MovingAvg"]=moving_avg
stock.head()
```

```python
outputfile = "WalmartStock.csv"
stock.to_csv(outputfile, header = True)
```
