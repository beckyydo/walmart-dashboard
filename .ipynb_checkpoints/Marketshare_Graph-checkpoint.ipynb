```python
import pandas as pd
```

```python
share = pd.read_csv("Walmart_Marketshare_data.csv")
```

```python
state = pd.read_excel("State-Abbr.xlsx", dtype = object)
```

```python
state.head()
```

```python
state_df = share.merge(state, how = "left", on = "ABBR")
state_df = state_df.rename(columns={'FIPS':'STATE_CODE'})
state_df['STATE_CODE'] = [str(code) for code in state_df['STATE_CODE']]
```

```python
df = pd.read_csv("US-Lat-Lon.csv", sep=";")
df.head()
```

```python
df = df.rename(columns={'City':'CITY', 'State':'ABR'})
df.head()
```

```python
graph = state_df.merge(df, how="inner", on = ['CITY','ABR'])
graph.head()
```

```python
graph['MARKET_SHARE'] = [ int(row[:-1]) for row in graph['MARKET_SHARE']]
graph.head()
```

```python
geocode = pd.read_excel("Geocodes.xlsx", dtype= object)
```

```python
geocode.columns = ['Summary','STATE_CODE','COUNTY','County_Subdivision','Place','Consolidated_City','Area_Name']
```

```python
clean_name = []
for name in geocode['Area_Name']:
    if (name.count(" ") != 0):
        reverse = name[::-1].split(" ", 1)[1]
        clean_name.append(reverse[::-1])
    else:
        clean_name.append(name)
```

```python
geocode['CITY'] = clean_name
geocode = geocode[['STATE_CODE','COUNTY','CITY']]
```

```python
geocode.head()
```

```python
share_df = state_df.merge(geocode, how="inner", on =['CITY','STATE_CODE'])
```

```python
share_df['FIPS'] = share_df['STATE_CODE'] + share_df['COUNTY']
share_df['MARKET_SHARE'] = [int(share.replace("%","")) for share in share_df['MARKET_SHARE']]
```

```python
import plotly.figure_factory as ff

import numpy as np


colorscale = ["#BA2607","#CF6852","#6baed6","#2171b5","#08306b"]

endpts = [60,70,80,90]
fips = share_df['FIPS']
values = share_df['MARKET_SHARE']

fig = ff.create_choropleth(
    fips = fips, 
    values = values,
    binning_endpoints = endpts,
    colorscale = colorscale,
    show_state_data = True,
    show_hover = True, 
    #centroid_marker={'opacity': 0},
    #asp=2.9, 
    title = 'Walmart Marketshare Over USA (%)',
    legend_title = 'Marketshare Percentage (%)'
)

fig.layout.template = None
fig.show()
```
