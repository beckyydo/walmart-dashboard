```python
import pandas as pd
```

```python
share = pd.read_csv("Walmart_Marketshare_data.csv")
```

```python
share = share.rename(columns={'STATE':'ABBR'})
```

```python
state = pd.read_excel("State-Abbr.xlsx", dtype = object)
```

```python
state.head()
```

```python
state_df = share.merge(state, how = "left", on = "ABBR")
state_df = state_df.rename(columns={'FIPS':'STATE_CODE'})
state_df['STATE_CODE'] = [str(code) for code in state_df['STATE_CODE']]
```

```python
state_df.head()
```

```python
geocode = pd.read_excel("Geocodes.xlsx", dtype= object)
```

```python
geocode.columns = ['Summary','STATE_CODE','County','County_Subdivision','Place','Consolidated_City','Area_Name']
```

```python
clean_name = []
for name in geocode['Area_Name']:
    if (name.count(" ") != 0):
        reverse = name[::-1].split(" ", 1)[1]
        clean_name.append(reverse[::-1])
    else:
        clean_name.append(name)
```

```python
geocode['CITY'] = clean_name
```

```python
geocode.head()
```

```python
share_df = state_df.merge(geocode, how="left", on =['CITY','STATE_CODE'])
```
